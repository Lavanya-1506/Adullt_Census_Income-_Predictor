```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/raw/adult.csv")
```

```python
df.replace(" ?", np.nan, inplace=True)

df.dropna(inplace=True)
```

```python
df["age_group"] = pd.cut(
    df["age"],
    bins=[0,25,45,65,100],
    labels=[
        "Young",
        "Adult",
        "Middle",
        "Senior"
    ]
)
```

```python
df["capital_net"] = (
    df["capital.gain"]
    -
    df["capital.loss"]
)
```

```python
df["work_intensity"] = pd.cut(
    df["hours.per.week"],
    bins=[0,20,40,60,100],
    labels=[
        "Low",
        "Normal",
        "High",
        "Extreme"
    ]
)
```

```python
df.to_csv(
    "../data/processed/cleaned_data.csv",
    index=False
)
```
